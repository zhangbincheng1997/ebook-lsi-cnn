```python
with open('me_train.json', 'r') as f:
    data = json.load(f)
```

```python
webqa = []
out = ''
for key in list(data.keys()):
    d = {}
    c = []
    item_id = int(key[6:]) # item_id
    question = data[key]['question']
    evidences = data[key]['evidences']
    for ekey in list(evidences.keys()):
        passage_id = int(key[6:] + ekey[13:]) # passage_id
        content = evidences[ekey]['evidence']
        label = 0 if evidences[ekey]['answer'][0] == 'no_answer' else 1
        dd = {}
        dd['passage_id'] = passage_id
        dd['content'] = content
        dd['label'] = label
        c.append(dd)
        out += str(passage_id) + ',' + str(label) + '\n'
    d['item_id'] = item_id
    d['question'] = question
    d['passages'] = c
    webqa.append(d)
```

```python
webqa[0]
```

```python
with open('WebQA.json', 'w') as f:
    f.write(json.dumps(webqa, ensure_ascii=False))
```

```python
with open('WebQA_label.txt', 'w') as f:
    f.write(out)
```
